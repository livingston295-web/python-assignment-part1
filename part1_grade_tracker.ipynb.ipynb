```python

# part 1PYTHON BASICS & CONTROL FLOW
# Theme: Student Grade Tracker


# Task 1 - Data Parsing & Profile Cleaning


print("========== TASK 1 ==========\n")

raw_students = [
    {"name": " ayesha SHARMA ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma", "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": " Priya Nair ", "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA", "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ", "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []

for student in raw_students:
    clean_name = student["name"].strip().title()
    clean_roll = int(student["roll"])
    clean_marks = [int(mark) for mark in student["marks_str"].split(", ")]

    cleaned_student = {
        "name": clean_name,
        "roll": clean_roll,
        "marks": clean_marks
    }

    cleaned_students.append(cleaned_student)

    # Check if every word in the name contains only alphabets
    if all(word.isalpha() for word in clean_name.split()):
        name_status = "✓ Valid name"
    else:
        name_status = "✗ Invalid name"

    print("=" * 32)
    print(f"Student : {clean_name}")
    print(f"Roll No : {clean_roll}")
    print(f"Marks    : {clean_marks}")
    print(name_status)
    print("=" * 32)
    print()

# Print name in ALL CAPS and lowercase for roll no 103
for student in cleaned_students:
    if student["roll"] == 103:
        print("Roll Number 103 Name Formats:")
        print("ALL CAPS :", student["name"].upper())
        print("lowercase:", student["name"].lower())
        print()


# Task 2 - Marks Analysis Using Loops & Conditionals

print("\n========== TASK 2 ==========\n")

student_name = "Ayesha Sharma"
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]

print(f"Student Name: {student_name}\n")

for i in range(len(subjects)):
    subject = subjects[i]
    mark = marks[i]

    if 90 <= mark <= 100:
        grade = "A+"
    elif 80 <= mark <= 89:
        grade = "A"
    elif 70 <= mark <= 79:
        grade = "B"
    elif 60 <= mark <= 69:
        grade = "C"
    else:
        grade = "F"

    print(f"{subject}: {mark} -> Grade {grade}")

total_marks = sum(marks)
average_marks = round(total_marks / len(marks), 2)

highest_mark = max(marks)
lowest_mark = min(marks)

highest_subject = subjects[marks.index(highest_mark)]
lowest_subject = subjects[marks.index(lowest_mark)]

print("\nSummary:")
print(f"Total Marks        : {total_marks}")
print(f"Average Marks      : {average_marks}")
print(f"Highest Subject    : {highest_subject} ({highest_mark})")
print(f"Lowest Subject     : {lowest_subject} ({lowest_mark})")

# While loop for marks-entry system
print("\n--- Marks Entry System ---")
new_subjects_added = 0

while True:
    subject_name = input("Enter subject name (or type 'done' to stop): ").strip()

    if subject_name.lower() == "done":
        break

    mark_input = input(f"Enter marks for {subject_name} (0-100): ").strip()

    try:
        mark_value = float(mark_input)

        if 0 <= mark_value <= 100:
            subjects.append(subject_name)
            marks.append(mark_value)
            new_subjects_added += 1
            print(f"{subject_name} added successfully.\n")
        else:
            print("Warning: Marks must be between 0 and 100.\n")

    except ValueError:
        print("Warning: Invalid marks input. Please enter a number.\n")

updated_average = round(sum(marks) / len(marks), 2)

print("\nUpdated Result:")
print(f"New Subjects Added : {new_subjects_added}")
print(f"Updated Average    : {updated_average}")


# Task 3 - Class Performance Summary

print("\n========== TASK 3 ==========\n")

class_data = [
    ("Ayesha Sharma", [88, 72, 95, 60, 78]),
    ("Rohit Verma", [55, 68, 49, 72, 61]),
    ("Priya Nair", [91, 85, 88, 94, 79]),
    ("Karan Mehta", [40, 55, 38, 62, 50]),
    ("Sneha Pillai", [75, 80, 70, 68, 85]),
]

pass_count = 0
fail_count = 0
topper_name = ""
topper_avg = 0
class_averages = []

print("Name | Average | Status")
print("-" * 40)

for name, student_marks in class_data:
    avg = round(sum(student_marks) / len(student_marks), 2)
    class_averages.append(avg)

    if avg >= 60:
        status = "Pass"
        pass_count += 1
    else:
        status = "Fail"
        fail_count += 1

    print(f"{name} | {avg:.2f} | {status}")

    if avg > topper_avg:
        topper_avg = avg
        topper_name = name

class_average = round(sum(class_averages) / len(class_averages), 2)

print("\nClass Summary:")
print(f"Passed Students : {pass_count}")
print(f"Failed Students : {fail_count}")
print(f"Class Topper    : {topper_name} ({topper_avg:.2f})")
print(f"Class Average   : {class_average:.2f}")


# Task 4 - String Manipulation Utility


print("\n========== TASK 4 ==========\n")

essay = " python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning. "

# Step 1
clean_essay = essay.strip()
print("1. Stripped Essay:")
print(clean_essay)
print()

# Step 2
print("2. Title Case:")
print(clean_essay.title())
print()

# Step 3
python_count = clean_essay.count("python")
print("3. Count of 'python':")
print(python_count)
print()

# Step 4
replaced_essay = clean_essay.replace("python", "Python 🐍")
print("4. Replaced Essay:")
print(replaced_essay)
print()

# Step 5
sentences = clean_essay.split(". ")
print("5. Sentence List:")
print(sentences)
print()

# Step 6
print("6. Numbered Sentences:")
for index, sentence in enumerate(sentences, start=1):
    if not sentence.endswith("."):
        sentence += "."
    print(f"{index}. {sentence}")
```
